```python
from PIL import Image
import pytesseract
import os

pytesseract.pytesseract.tesseract_cmd = r'C:\Users\dreynosoh\AppData\Local\Programs\Tesseract-OCR\tesseract.exe'
custom_config = r'--oem 3 --psm 6'

def extraer_texto_ocr(screens):
    screen = os.listdir(screens)
    mlms = []
    prefijos_buscados = ('Usado', 'usado', '(Usado)', '(usado)','(usado', 'usado)', '(Usado','Usado)', 'usado', 'Usad')
    for imag in screen:
        ruta_imagen = os.path.join(screens,imag)
        imagen = Image.open(ruta_imagen)
        texto = pytesseract.image_to_string(imagen, lang='spa', config=custom_config)
    
        palabras = texto.split()
        palabras_encontradas = [p for p in palabras if p.startswith(prefijos_buscados)]
        if palabras_encontradas:
            print(f'Encontradas palabras que empiezan con "usado" en {imag}:')
            mlms.append(imag)
            for palabra in palabras_encontradas:
                print('-', palabra)
            #print('\nTexto completo extraído:')
            #print(texto)
        else:
            print(f'No se encontraron palabras que empiecen con "usado" en {imag}. 👍')
            #print(texto)

    return mlms


def ref_extraer_texto_ocr(screens):
    screen = os.listdir(screens)
    mlms = []
    prefijos_buscados = ('repuesto','Repuesto','Refacción','Refaccion','refaccion','piezas','pzas','P/','p/','reemplazo','Reemplazo','accesorio','Accesorio','parte','complemento', 'Complemento','Poncho', 'poncho', 'batería', 'Batería','bateria')
    for imag in screen:
        ruta_imagen = os.path.join(screens,imag)
        imagen = Image.open(ruta_imagen)
        region_titulo = imagen.crop((600, 250, 1300, 550)) 
        texto = pytesseract.image_to_string(region_titulo, lang='spa', config=custom_config)
    
        palabras = texto.split()
        palabras_encontradas = [p for p in palabras if p.startswith(prefijos_buscados)]
        if palabras_encontradas:
            print(f'Encontradas palabras que empiezan con "usado" en {imag}:')
            mlms.append(imag)
            for palabra in palabras_encontradas:
                print('-', palabra)
            #print('\nTexto completo extraído:')
            #print(texto)
        else:
            print(f'No se encontraron palabras que empiecen con "refacciones" en {imag}. 👍')
            #print(texto)

    return mlms
    
if __name__ == "__main__":
    screens = r'C:\Users\dreynosoh\OneDrive - Truper, S.A. de C.V\Escritorio\Análisis MeLi\3. Screenshots P'
    mlms = extraer_texto_ocr(screens)
    mlms2 = ref_extraer_texto_ocr(screens)
```

```python
import pandas as pd 
import numpy as np

mlm = pd.DataFrame(mlms, columns=['MLM'])
mlm['#'] = range(1,len(mlm)+1)
mlm['MLM'] = mlm['MLM'].str[:-4] 
indice = mlm.pop('#')
mlm.insert(0,'#',indice)
#mlm.to_excel('MLM usados.xlsx',index = False)
display(mlm.head())

mlm2 = pd.DataFrame(mlms2, columns=['MLM'])
mlm2['MLM'] = mlm2['MLM'].str[:-4]
mlm2['#'] = np.arange(1.1,len(mlm2)+4,1.11)
indice2 = mlm2.pop('#')
mlm2.insert(0,'#',indice2)
#mlm2.to_excel('MLM refacciones.xlsx',index = False)
display(mlm2.head())

mlm3 = pd.concat([mlm,mlm2],ignore_index = True)
mlm3.to_excel('Usados y Reparados.xlsx', index = False)
```
